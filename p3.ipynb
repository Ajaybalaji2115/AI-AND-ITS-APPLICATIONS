```python
import re

def intelligent_calculator():
    print("--- AI Calculator ---")
    print("Type your request (e.g., 'multiply 10 and 5' or 'add 3 to 7')")
    print("Type 'exit' to quit.")

    while True:
        user_input = input("\nHow can I help? ").lower().strip()

        if user_input == 'exit':
            print("Goodbye!")
            break

        # Extract all numbers (integers or decimals) from the string
        numbers = [float(n) for n in re.findall(r"[-+]?\d*\.\d+|\d+", user_input)]

        if len(numbers) < 2:
            print("Error: I need at least two numbers to perform an operation.")
            continue

        # Define operation keywords
        if any(word in user_input for word in ['add', 'plus', 'sum', 'total']):
            result = sum(numbers)
            op = "sum"
        
        elif any(word in user_input for word in ['subtract', 'minus', 'difference', 'take away']):
            # Subtracts subsequent numbers from the first
            result = numbers[0] - sum(numbers[1:])
            op = "difference"

        elif any(word in user_input for word in ['multiply', 'times', 'product']):
            result = 1
            for n in numbers:
                result *= n
            op = "product"

        elif any(word in user_input for word in ['divide', 'divided by', 'split']):
            try:
                result = numbers[0]
                for n in numbers[1:]:
                    result /= n
                op = "quotient"
            except ZeroDivisionError:
                print("Error: Math says no. You cannot divide by zero.")
                continue
        
        else:
            print("Error: I couldn't identify the operation. Try using words like 'add', 'minus', 'times', or 'divide'.")
            continue

        # Display the result
        print(f"The {op} of {numbers} is: {result}")

if __name__ == "__main__":
    intelligent_calculator()
```
