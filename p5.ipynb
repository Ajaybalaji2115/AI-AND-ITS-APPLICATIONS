```python
#AJAY
import heapq
graph = {
    'A': ['B', 'D'],
    'B': ['E', 'G'],
    'D': ['G', 'H'],
    'G': ['I'],
    'H': ['I'],
    'E': [],
    'I': []
}
heuristic = {
    'A': 999,  
    'B': 5,
    'D': 2,
    'E': 4,
    'G': 6,
    'H': 1,
    'I': 0
}
edge_cost = 1 
def greedy_bfs(start, goal, graph, heuristic):
  
    priority_queue = [(heuristic[start], start)]
    visited = set()
    path = {start: []}  

    while priority_queue:
        current_h, current_node = heapq.heappop(priority_queue)

        if current_node == goal:
            full_path = path[current_node] + [current_node]
            print(f"Greedy Best-First Search Path: {full_path}")
            return full_path

        if current_node not in visited:
            visited.add(current_node)

            for neighbor in graph[current_node]:
                if neighbor not in visited:
                    heapq.heappush(priority_queue, (heuristic[neighbor], neighbor))
                    path[neighbor] = path[current_node] + [current_node]
    print("Greedy Best-First Search: No path found")
    return None
def a_star(start, goal, graph, heuristic, edge_cost):
    
    priority_queue = [(heuristic[start], 0, start)]
    g_scores = {node: float('inf') for node in graph}
    g_scores[start] = 0
    path = {start: []}

    while priority_queue:
        f_value, current_g, current_node = heapq.heappop(priority_queue)

        if current_node == goal:
            full_path = path[current_node] + [current_node]
            print(f"A* Search Path: {full_path}")
            return full_path

        for neighbor in graph[current_node]:
            tentative_g_score = current_g + edge_cost

            if tentative_g_score < g_scores[neighbor]:
                g_scores[neighbor] = tentative_g_score
                f_value = tentative_g_score + heuristic[neighbor]
                heapq.heappush(priority_queue, (f_value, tentative_g_score, neighbor))
                path[neighbor] = path[current_node] + [current_node]

    print("A* Search: No path found")
    return None

start_node = 'A'
goal_node = 'I'

greedy_bfs(start_node, goal_node, graph, heuristic)
a_star(start_node, goal_node, graph, heuristic, edge_cost)
```
